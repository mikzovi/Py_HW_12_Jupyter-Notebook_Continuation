Прочитайте данные о продажах мыла для купания за все месяцы и покажите их с помощью столбчатой диаграммы.

Рассчитайте общие данные о продажах за год для каждого продукта и покажите их с помощью круговой диаграммы.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
root = os.getcwd()
filename = os.path.join(root, 'company_sales_data.csv')
```

```python
data = pd.read_csv(filename)
data.head()
```

```python
months = data['month_number'].tolist()
facecream = data['facecream'].tolist()
facewash = data['facewash'].tolist()
toothpaste = data['toothpaste'].tolist()
bathingsoap = data['bathingsoap'].tolist()
shampoo = data['shampoo'].tolist()
moisturizer = data['moisturizer'].tolist()
```

# 1. Данные о продажах:

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1,1,1)
ax.plot(months,facecream, label = 'Крем для лица')
ax.plot(months,facewash, label = 'Средство для умывания')
ax.plot(months,toothpaste, label = 'Зубная паста')
ax.plot(months,bathingsoap, label = 'Мыло для купания')
ax.plot(months,shampoo, label = 'Шампунь')
ax.plot(months,moisturizer, label = 'Увлажняющий крем')
plt.title('Объемы продаж по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Объем продаж', labelpad=20);
plt.xticks(months)
plt.yticks([i *1000 for i in range(20)])
plt.legend(loc = 'upper left')
plt.grid(linewidth = 1, linestyle = '--')
plt.show()
```

# 2. Данные о продажах мыла для купания

```python
plt.figure(figsize=(10,5))
plt.bar(months,bathingsoap)
plt.title('Объемы продаж мыла по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Объем продаж', labelpad=20);
plt.xticks(months)
plt.yticks([i *1000 for i in range(20)])
plt.show()
```

# 3. Общие данные о продажах за год для каждого продукта

```python
labels = 'Крем для лица', 'Средство для умывания', 'Зубная паста', 'Мыло для купания', 'Шампунь', 'Увлажняющий крем'

sizes = [
    sum(facecream),
    sum(facewash),
    sum(toothpaste),
    sum(bathingsoap),
    sum(shampoo),
    sum(moisturizer)
]
explode = (0, 0.1, 0, 0, 0 ,0)
plt.figure(figsize=(10,5))
plt.title('Объемы продаж товаров за год')

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=100)

plt.show()
```
